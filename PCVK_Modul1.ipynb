```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from skimage import io
from skimage import transform
from PIL import Image

```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://imgx.parapuan.co/crop/0x36:1105x798/x/photo/2022/02/26/istock-468382096jpg-20220226021318.jpg"]

# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)    #membaca image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #mengubah ukuran menjadi 1/2 lebih kecil
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #mengubah warna ke rgb
  final_frame = cv.hconcat((image,image_2))  #menggabungkan image
  cv2_imshow(final_frame)  #show image
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image : tinggi x lebar ", tinggi," x ", lebar)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.line(image_3, (0,0), (lebar, tinggi), (255,255,255))
cv.line(image_3, (lebar,0), (0, tinggi), (255,255,255))

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
