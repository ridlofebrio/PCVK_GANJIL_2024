2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
import cv2 as cv
import matplotlib.pyplot as plt


img = cv.imread('/content/drive/MyDrive/image/kucing1.jpeg')

# salinan untuk setiap modifikasi
img_rb = img.copy()
img_gb = img.copy()

# Hapus channel Green (GB dihapus)
img_rb[:, :, 1] = 0  # Kanal 1 adalah Green

# Hapus channel Red (R dihapus)
img_gb[:, :, 2] = 0  # Kanal 2 adalah Red

# Konversi dari BGR ke RGB untuk ditampilkan dengan matplotlib
img_rb_rgb = cv.cvtColor(img_rb, cv.COLOR_BGR2RGB)
img_gb_rgb = cv.cvtColor(img_gb, cv.COLOR_BGR2RGB)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rb_rgb)
plt.title('Red-Blue Channel')

plt.subplot(1, 2, 2)
plt.imshow(img_gb_rgb)
plt.title('Green-Blue Channel')

plt.show()
```

7. Buat rectangle dan circle pada bagian wajah dari image foto anda saat
beraktifitas (bukan pasfoto).


```python
urls = ["https://www.rollingstone.com/wp-content/uploads/2024/09/Elon-Musk.jpg?w=1581&h=1054&crop=1",]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  cv.circle(image_2, center=(540,220), radius=120, color=(0,100,0), thickness = 2)
  cv.rectangle(image_2, pt1=(400,100), pt2=(720,350), color=(0,255,0), thickness = 2 )
  cv2_imshow(image_2)


```
